<h1 style="height:50px; width:50px; color:red";>Gender</h1>

```python
ax=sns.countplot(x="Gender",data=df)
```

```python
ax=sns.countplot(data=df,x="Gender")
for bar in ax.containers:
    ax.bar_label(bar)
```

<h1 style="height:100px; width:100px; color:red";>Gender vs Amount</h1>

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_Gen=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
#color=['yellow' if g=="M" else "black" for g in sales_Gen["Gender"]]
sns.barplot(x="Gender",y="Amount",hue="Gender",data=sales_Gen)
plt.show()
```

from above graphh we can see that most of the buyers are female and power of buying from female is more than male

```python
df.columns
```

```python
order=sorted(df["Age Group"].unique())
sns.countplot(x="Age Group",hue="Gender",data=df,order=order,palette={"M":"red","F":"blue"})
plt.show()
```

```python
asi=sns.countplot(x="Age Group",data=df,hue="Gender",palette={"M":"red","F":"green"},order=sorted(df["Age Group"].unique()))
for bar in asi.containers:
    asi.bar_label(bar)
```

```python
df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sales_Age=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=True)
sns.barplot(x="Age Group",y="Amount",data=sales_Age,hue="Age Group")
plt.show()
```

we can see that most of the buyers from age group 26 to 35

<h1 style="height:50px; width:50px; color:red";>State</h1>

```python
    df["State"]
```

```python
order=sorted(df["State"].unique())
plt.figure(figsize=(15,5))
sns.countplot(x="State",data=df,order=order,hue="State")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

```python
sales_state=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=True)
plt.figure(figsize=(15,5))
sns.barplot(x="State",y="Orders",data=sales_state,hue="State")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

```python
sales_state=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(style="whitegrid") 
plt.figure(figsize=(15,5))
sns.barplot(x="State",y="Amount",data=sales_state,hue="State")
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

*we can conclude that Uttar pradesh purchasing rate is more and Telangana as least purchasing rate overall*

<h1 style="height:100px; width:100px; color:red";>Marital status</h1>

```python
ax=sns.countplot(x="Married?",data=df,hue="Married?")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sales_state=df.groupby(["Married?","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False) 
plt.figure(figsize=(15,5))
sns.barplot(x="Married?",y="Amount",data=sales_state,hue="Gender")
plt.xticks(rotation=0,ha="right")
plt.tight_layout()
plt.show()
```

*In the above cell we can see that female who not married paid more than female who married and same thing for mens also

```python
first_10_category=df['Product_Category'].unique()[:15]
plt.figure(figsize=(7,5))
sns.countplot(x='Product_Category',data=df,order=first_10_category,hue='Product_Category')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sales_occupation=df.groupby(['Product_Category'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='Product_Category',y="Amount",data=sales_occupation,hue='Product_Category')
plt.xticks(rotation=45,ha="right")
plt.tight_layout()
plt.show()
```

*from above graph we can tell that food category product as saled and reached highest than others*
